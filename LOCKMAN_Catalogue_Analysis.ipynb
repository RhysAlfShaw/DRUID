```python
import pandas as pd
import numpy as np

#PATH = '/data/typhon2/Rhys/data/LoFAR/Lockman_hole/facets/Topcat_matched_5arc.fits'
PATH_DRUID_MATCHED_TODR1 = '/data/typhon2/Rhys/data/LoFAR/Lockman_hole/DRUID_Filtered_mATCHED_TODR1_new.fits'
PATH = '/data/typhon2/Rhys/data/LoFAR/Lockman_hole/DRUID_DR1_Matched_to_HiResCat_new.fits'
Test_Path = '/data/typhon2/Rhys/data/LoFAR/Lockman_hole/facets/facet_01_catalogue.fits'
PATH_Given5Sigma_Cat = '/data/typhon2/Rhys/data/LoFAR/Lockman_hole/catalogue_filtered_full_SNR5_fluxscaled_withoffset_duplicatesfiltered.fits'

# load the fits table
from astropy.table import Table

table = Table.read(PATH).to_pandas()
table.columns
```

```python
# plot flux histogram
import matplotlib.pyplot as plt

bins = np.logspace(-5,0,1000)

Flux_BSFval = np.histogram(table['Isl_Total_flux'],bins=bins)[0]
Flux_DRUIDval = np.histogram(table['Flux_total'],bins=bins)[0]
Flux_peak_BSFval = np.histogram(table['Peak_flux_1'],bins=bins)[0]

plt.step(bins[:-1],Flux_BSFval,label='BSF',marker='')
plt.step(bins[:-1],Flux_DRUIDval,label='DRUID',marker='')

plt.xscale('log')

plt.legend()
plt.xlabel('Flux (Jy)')
plt.ylabel('$N$')
plt.show()
```

```python
table['flux_total_err'] = table['flux_total_err'] + (table['Noise'])
table['flux_peak_err'] = table['flux_peak_err'] + (table['Noise'])
```

```python
# calculate the difference between the two fluxes
table['diff_0'] = abs(table['Flux_peak'] - table['Peak_flux_2'])/abs(table['Flux_peak'])
table['diff_1'] = abs(table['Flux_peak'] - table['Peak_flux_2'])/abs(table['Peak_flux_2'])

print('Number of sources: ',len(table))
# if abs(diff) > 1e-6, then remove from the table
print(table['diff_0'].describe())
diff_factor = 1.2
table = table[table['diff_0'] < diff_factor]
table = table[table['diff_1'] < diff_factor]
print('Number of sources: ',len(table))
```

```python

import matplotlib.pyplot as plt
#plt.style.use('/home/rs17612/GitHub/mplrc_sotiria.mplstyle')

flux_scale = 1

x = np.linspace(0, 2, 1000)
y = x

fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].errorbar(x = table['Isl_Total_flux'], y = table['Flux_total']*flux_scale,xerr=table['E_Isl_Total_flux'],yerr=table['flux_total_err'],color='crimson', 
               alpha=0.05,zorder=10,label='Isl',linestyle='',marker='o')

#ax[0].errorbar(x=table['Total_flux_2'], y=table['Flux_total']*flux_scale,xerr=table['E_Total_flux_2'],
#               yerr=table['flux_total_err'], alpha=0.05,zorder=10,label='Total',linestyle='',marker='o')

ax[0].plot(x, y, color='black', linestyle='--',linewidth=1,marker='',zorder=1)
ax[0].set_ylabel('DRUID Flux density (Jy)')
ax[0].set_xlabel('LCatalogue Flux density (Jy)')
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].set_xlim(1e-4,1)

ax[1].errorbar(x=table['Peak_flux_2'], y=table['Flux_peak']*flux_scale,xerr=table['E_Peak_flux_2'],yerr=table['flux_peak_err'],
               color='crimson', alpha=0.1,zorder=10,linestyle='',marker='o')

ax[1].plot(x, y, color='black', linestyle='--', linewidth=1, marker='', zorder=1)
ax[1].set_ylabel('DRUID Flux density  (Jy)') 
ax[1].set_xlabel('Catalogue Flux density (Jy)')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].set_xlim(1e-4,1)

# Add legends
legend1 = ax[0].legend()
legend2 = ax[1].legend()

# Set the alpha value of the legend markers to 1
for legend_handle in legend1.legendHandles:
    legend_handle.set_alpha(1)

for legend_handle in legend2.legendHandles:
    legend_handle.set_alpha(1)
plt.show()
```

```python
# plot the difference in flux as a function of SNR
import matplotlib.pyplot as plt
table['diff_0'] = (table['Flux_total'] - table['Isl_Total_flux'])/(table['Isl_Total_flux'])
table['diff_1'] = (table['Flux_total']/table['Corr_f'] - table['Isl_Total_flux'])/(table['Isl_Total_flux'])

#plt.style.use('/home/rs17612/GitHub/mplrc_sotiria.mplstyle')
plt.scatter(table['SNR'],table['diff_0'],color='crimson',alpha=0.1)
plt.scatter(table['SNR'],table['diff_1'],color='blue',marker='x',alpha=0.1)
plt.ylabel('Fractional difference between measured and catalogue flux')
plt.xlabel('SNR')
plt.ylim(-1,1)
plt.xscale('log')
plt.show()
```

```python
from matplotlib.patches import Ellipse
import matplotlib.pyplot as plt


# run DRUID on this image   
# define colors for the contours
colors = ['red','orange','purple','pink','green']
# set the colors to the matplotlib default
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=colors) 
from DRUID import sf
from astropy.wcs import WCS
# issue here with mode=Radio and the added padding and aprseing image, not issue with imagepath. Creating spurious sources?
findmysources = sf(image=image_25, image_path=None,
                   pb_path=None, mode='Radio',
                   cutup=False, cutup_size=1000,cutup_buff=0,
                   output=False,GPU=True,header=header)

findmysources.set_background(detection_threshold=5,analysis_threshold=2.1,mode='mad_std',
                             bg_map_bool=False, box_size=None)

findmysources.phsf(lifetime_limit_fraction=1.5)
findmysources.source_characterising(use_gpu=True)
print(findmysources.catalogue)
#wcs projection
wcs = WCS(header).celestial

# do the plot below with projection)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=wcs)

# plot the image with the sources
ax.imshow(image_25,origin='lower',cmap='Greys',vmin=0,vmax=0.001)
# plot contours
contours = findmysources.catalogue['contour']
for con in contours:
    plt.plot(con[:,1],con[:,0])
# plot the gaussian compoent on top
#Elli = Ellipse((proviede_cat['X']-5,proviede_cat['Y']),width=proviede_cat['Majp']*2,height=proviede_cat['Minp']*2,angle=proviede_cat['PA']-90,fill=False,edgecolor='red',linewidth=2,linestyle='--')
#ax.add_artist(Elli)
plt.xlabel('Right Ascension')
plt.ylabel('Declination')
plt.savefig('figures/DRUID_contours_RLAGN_Lockman.svg')
plt.show()
```
